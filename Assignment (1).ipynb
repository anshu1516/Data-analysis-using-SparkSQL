```
%sql
DROP TABLE IF EXISTS ipl_ball_by_ball;

CREATE TABLE ipl_ball_by_ball USING DELTA LOCATION '/delta/ipl_ball_by_ball/'
```

```
%sql
select bowling_team,max(extra_runs) as highest_extra_runs from ipl_ball_by_ball group by bowling_team order by highest_extra_runs desc
```

```
%sql
select bowler,sum(is_wicket) as leading_wicket_taker from ipl_ball_by_ball where dismissal_kind != 'run out' group by bowler order by leading_wicket_taker desc
```

```
%sql
select b.batting_team,sum(total_runs) as highest_run_scorer from ipl_ball_by_ball b inner join ipl_matches m on b.match_id=m.match_id where m.method = 'D/L' group by b.batting_team order by highest_run_scorer desc
```

```
%sql
select batsman,(sum(batsman_runs)/sum(ball))*100 as strike_rate from ipl_ball_by_ball where extras_type not in('wides','noballs') and overs between 
1 and 6 group by batsman order by strike_rate desc
```

```
%sql
select v.venue,max(extra_runs) as venue_highest_extra_runs from ipl_matches m inner join ipl_venue v on v.venue_id=m.venue_id 
inner join ipl_ball_by_ball b on m.match_id=b.match_id group by v.venue order by venue_highest_extra_runs desc
```

```
%sql
select player_of_match,count(player_of_match) as most_number_of_pom from ipl_matches where neutral_venue = 1 group by player_of_match order by most_number_of_pom desc
```

```
%sql
select batsman,sum(batsman_runs)/sum(is_wicket) as batting_average from ipl_ball_by_ball group by batsman 
order by batting_average desc limit 10
```

```
%sql
(select umpire1 as umpire,count(umpire1) as most_officiated from ipl_matches group by umpire1 order by most_officiated desc) union (select umpire2 as umpire,count(umpire2) as most_officiated from ipl_matches group by umpire2)
```

```
%sql
select venue,city,sum(batsman_runs) as highest_individual_score from ipl_matches inner join ipl_venue on ipl_matches.venue_id=ipl_venue.venue_id inner join ipl_ball_by_ball on ipl_matches.match_id=ipl_ball_by_ball.match_id where batsman='V Kohli' group by venue,city order by highest_individual_score desc
```
